Lekte runt lite med koden för kårhuset och partikelrörelse

```python
#------------------------CREATE ARENA----------------------#
import math
import numpy as np 
from matplotlib import pyplot as plt
from tkinter import Tk, Canvas
import time

L = 140
N_particles = 100
speed = 50
dt = 0.01

# Väggarna
x1 = [0, 0, 0, 12]
y1 = [9, 28, 28, 28]

x2 = [17, 30, 30, 30, 30, 104]
y2 = [28, 28, 28, 40, 40, 40]

x3 = [109, 140, 140, 140]
y3 = [40, 40, 40, 10]

x4 = [140, 140, 140, 104, 104, 104, 104, 67, 67, 67, 67, 58, 58, 58, 58, 44, 44, 44, 44, 0]
y4 = [5, 0, 0, 0, 0, 30, 30, 30, 30, 6, 6, 6, 6, 20, 20, 20, 20, 9, 9, 9]

wall_thickness = 3
scale = 6 # Scale factor for better visualization in the canvas

# Canvas size
canvas_size = 1000
max_y = max(max(y1), max(y2), max(y3), max(y4))  # Find the maximum Y-coordinate for flipping


window = tk.Tk()
window.title("Test")
canvas = tk.Canvas(window, width=canvas_size, height=canvas_size, bg="white")
canvas.pack()

def draw_walls(x_coords, y_coords, thickness, scale_factor):
    for i in range(len(x_coords) - 1):
        x_start = x_coords[i] * scale_factor + 50
        y_start = (max_y - y_coords[i]) * scale_factor + 200  # Flip Y-axis
        x_end = x_coords[i + 1] * scale_factor + 50
        y_end = (max_y - y_coords[i + 1]) * scale_factor + 200  # Flip Y-axis

        # Draw a rectangle for each segment to represent wall thickness
        canvas.create_rectangle(
            x_start,
            y_start,
            x_end + thickness if x_start == x_end else x_end,
            y_end + thickness if y_start == y_end else y_end,
            fill="black",
        )

x = np.random.uniform(0, L, N_particles)
y = np.random.uniform(0, L, N_particles)
particles = []

for j in range(N_particles):
    particles.append(
        canvas.create_oval(
            x[j] * scale - 2,
            y[j] * scale - 2,
            x[j] * scale + 2,
            y[j] * scale + 2,
            fill="#00C0C0"))

#Testar en punkt i en dörr till att börja med
door_1_x = 21 * scale #21-25
door_1_y = 46 * scale 

# Rör sig mot dörrarna, men behöver se till att de slår mot väggarna och varandra - fixar detta i en ny fil:D
def move_toward_target(px, py):
    dx = door_1_x-px
    dy = door_1_y-py
    dist = np.sqrt(dx**2 + dy**2)
    return (speed * dx / dist, speed * dy / dist) if dist > 0 else (0, 0)

def update_particles():
    global x, y
    for i in range(N_particles):
        vx, vy = move_toward_target(x[i] * scale, y[i] * scale)
        x[i] += vx * dt / scale
        y[i] += vy * dt / scale

        canvas.coords(
            particles[i],
            x[i] * scale - 2,
            y[i] * scale - 2,
            x[i] * scale + 2,
            y[i] * scale + 2)

def simulation():
    for _ in range(1000):
        update_particles()
        window.update()
        time.sleep(0.001)

draw_walls(x1, y1, wall_thickness, scale)
draw_walls(x2, y2, wall_thickness, scale)
draw_walls(x3, y3, wall_thickness, scale)
draw_walls(x4, y4, wall_thickness, scale)

simulation()
window.mainloop()
```
