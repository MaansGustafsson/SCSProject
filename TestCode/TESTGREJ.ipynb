```python
import math
import numpy as np 
from matplotlib import pyplot as plt
```

```python
N = 10

Lx = 160
Ly = 80

v0 = 600
dt = 0.01

t_tot = 50000
```

```python
#initialize

x = np.random.randint(low = 30, high = 70, size=[N])
y = np.random.randint(low = 30, high = 70, size=[N])
phi = np.random.rand(N)*2*np.pi
vx = v0*np.cos(phi)
vy = v0*np.sin(phi)

x_path = np.zeros([N,t_tot])
y_path = np.zeros([N,t_tot])
```

```python
for t in range(t_tot):

    x_path[:,t] = x
    y_path[:,t] = y

    nx = x + vx*dt
    ny = y + vy*dt
    nvx = vx
    nvy = vy

    for i in range(N):

        if nx[i] < 0:
            x[i] = -nx[i]
            vx[i] = -nvx[i]

        elif nx[i] > Lx:
            x[i] = Lx - (nx[i]-Lx)
            vx[i] = -nvx[i]
        else:
            x[i] = nx[i]
            vx[i] = nvx[i]

        if ny[i] < 0:
            y[i] = -ny[i]
            vy[i] = -nvy[i]

        elif ny[i] > Ly:
            y[i] = Ly - (ny[i]-Ly)
            vy[i] = -nvy[i]
        else:
            y[i] = ny[i]
            vy[i] = nvy[i]

        if nx[i] < 20 and ny[i] > 60:
            y[i] = 120-ny[i]
            x[i] = 40-nx[i]
            vy[i] = -nvy[i]
            vx[i] = -nvx[i]
        
        if nx[i] > 120 and ny[i] > 70:
            y[i] = 140-ny[i]
            x[i] = 240-nx[i]
            vy[i] = -nvy[i]
            vx[i] = -nvx[i]

        if nx[i] > 60 and nx[i] < 80 and ny[i] > 15 and ny[i] < 25:
            if ny[i] < 20:
                y[i] = 30-ny[i]
            else:
                y[i] = 50-ny[i]

            if nx[i] < 70:
                x[i] = 120-nx[i]
            else:
                x[i] = 160-nx[i]

            vy[i] = -nvy[i]
            vx[i] = -nvx[i]           

plt.scatter(x_path, y_path)
plt.xlim(0,Lx)
plt.ylim(0,Ly)


    


```
